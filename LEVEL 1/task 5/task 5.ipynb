```python
import numpy as np
import pandas as pd

# Example dataset
data = pd.DataFrame({
    'id': range(1, 101),
    'value': np.random.randint(1, 100, size=100)
})

# Simple random sampling
sample_size = 10
random_sample = data.sample(n=sample_size, random_state=42)

# Print the random sample
print("Simple Random Sample:")
print(random_sample)
```
