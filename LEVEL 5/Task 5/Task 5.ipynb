```python
from dataclasses import dataclass
import pandas as pd
import matplotlib.pyplot as plt
@dataclass
class DatasetStats:
    mean: float
    max: float
    min: float
    median: float
    def plot_histogram(self, data, column):
        plt.figure(figsize=(8, 5))
        plt.hist(data[column], bins=10, edgecolor='black')
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
houses = pd.read_csv(r'C:\Users\Welcome\OneDrive\Pictures\Desktop\LEVEL 5\Task 4\task 5 dataset housing.txt')
stats = DatasetStats(
    mean=houses['price'].mean(),
    max=houses['price'].max(),
    min=houses['price'].min(),
    median=houses['price'].median()
)
while True:
    print("\n=== Data Dashboard ===")
    print("1. Show statistics")
    print("2. Plot price histogram")
    print("3. Exit")
    choice = input("Enter your choice (1-3): ")
    if choice == '1':
        print("\nPrice Statistics:")
        print(f"Mean: {stats.mean:.2f}")
        print(f"Max: {stats.max:.2f}")
        print(f"Min: {stats.min:.2f}")
        print(f"Median: {stats.median:.2f}")
    elif choice == '2':
        stats.plot_histogram(houses, 'price')
    elif choice == '3':
        print("Exiting dashboard...")
        break
    else:
        print("Invalid choice. Please try again.")
```
