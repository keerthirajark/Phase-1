```python
import pandas as pd
import numpy as np

# Example dataset
data = pd.DataFrame({
    'x1': np.random.rand(100),
    'x2': np.random.rand(100) * 2,
    'x3': np.random.rand(100) * 3
})

# Calculate correlation matrix
correlation_matrix = data.corr()
print("Correlation Matrix:")
print(correlation_matrix)
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot between x1 and x2
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['x1'], y=data['x2'])
plt.title("Scatter Plot: x1 vs x2")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
# Pair plot for all variables
sns.pairplot(data)
plt.suptitle("Pair Plot of Variables", y=1.02)
plt.show()
# Heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
# Regression plot between x1 and x2
sns.regplot(x=data['x1'], y=data['x2'])
plt.title("Regression Plot: x1 vs x2")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
# Example dataset with a categorical variable
data['category'] = np.random.choice(['A', 'B', 'C'], size=100)

# Grouped bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='category', y='x1', data=data, ci=None)
plt.title("Grouped Bar Plot: x1 by Category")
plt.xlabel("Category")
plt.ylabel("x1")
plt.show()
# Box plot of x1 by category
sns.boxplot(x='category', y='x1', data=data)
plt.title("Box Plot: x1 by Category")
plt.xlabel("Category")
plt.ylabel("x1")
plt.show()
from pandas.plotting import scatter_matrix

# Scatter matrix for all variables
scatter_matrix(data, figsize=(10, 8), diagonal='kde')
plt.suptitle("Scatter Matrix of Variables", y=1.02)
plt.show()
# Example dataset with another categorical variable
data['category2'] = np.random.choice(['X', 'Y'], size=100)

# Cross-tabulation
cross_tab = pd.crosstab(data['category'], data['category2'])
print("Cross-Tabulation:")
print(cross_tab)
from scipy.stats import chi2_contingency

# Chi-Square test
chi2, p, dof, expected = chi2_contingency(cross_tab)
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
```
