```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic = pd.read_csv(r'C:\Users\Welcome\OneDrive\Pictures\Desktop\LEVEL 5\Task 2\Task 2 Titanic-Dataset.csv')
summary_stats = titanic[['Age', 'Fare']].describe()
survival_by_class = titanic.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(8, 5))
sns.barplot(x=survival_by_class.index, y=survival_by_class.values)
plt.title('Survival Rates by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class'])  # Label classes clearly
plt.show()
print("Summary Statistics:")
print(summary_stats)
print("\nSurvival Rates by Class:")
print(survival_by_class)
survival_by_gender = titanic.groupby('Sex')['Survived'].mean()
print("\nSurvival Rates by Gender:")
print(survival_by_gender)
```
